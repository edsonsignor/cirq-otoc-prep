## What is Cirq?

Cirq is a Python framework for building, simulating, transforming, and eventually running quantum circuits on supported backends.

## What is an OTOC?

An OTOC measures operator growth and scrambling through a non-time-ordered correlator such as `F(t) = < W_dagger(t) V_dagger W(t) V >`.

## How would I implement a toy OTOC in Cirq?

I would build `U`, insert a local perturbation `W`, apply `U_dagger`, measure a return signal, and sweep perturbation strength, depth, and noise.

## How do I know whether gates are good?

I would check gate fidelity, readout error, decoherence, leakage, crosstalk, calibration drift, qubit layout, and control circuits.
